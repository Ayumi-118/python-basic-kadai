```python
def find_max(numbers):
  max_num = numbers[0]
  for num in  numbers[1:]:
    if num > max_num:
      max_num = num
  return max_num
```

```python
class Product:
  def __init__(self,name:str,price:int,category:str,stock:int,colors:list,related_products:dict):
    self.name = name
    self.price = price
    self.category = category
    self.stock = stock
    self.colors = colors
    self.related_products = related_products

  def get_product_info(self):
    return f"【商品情報】商品名： {self.name} 価格：{self.price} カテゴリ:{self.category} 在庫数：{self.stock} 色：{self.colors} 関連商品：{self.related_products}"

  def update_product_info(self,name:str = None,price:int = None,category:str = None,stock:int = None,colors:list = None,related_products:dict = None):
    if name != None:
     self.name = name
    if price != None:
     self.price = price
    if category != None:
     self.category = category
    if stock != None:
     self.stock = stock
    if colors != None:
     self.colors = colors
    if related_products != None:
     self.related_products = related_products

  def show_related_products(self):
    high_related_products = []

    for name,score in self.related_products.items():
      if score>=5:
        high_related_products.append(name)

    return f"よく一緒に購入されている商品は{','.join(high_related_products)}です"
```
