```python
#上辺の長さ
upper_base = 10
#下辺の長さ
lower_base = 20
#高さ
height = 5

#台形の面積を計算
area = (upper_base + lower_base)*height/2

# 面積を整数で表示し、単位を付ける
print(f"{int(area)}cm²")
```
